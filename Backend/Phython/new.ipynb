```python
def calculator(num1, num2, operator):
    if operator == "+":
        return num1 + num2
    elif operator == "-":
        return num1 - num2
    elif operator == "*":
        return num1 * num2
    elif operator == "/":
        if num2 == 0:
            return "Error: cannot divide by Zero"
        return num1 / num2
    else:
        return "Invalid operator"
    
number1 = float(input("Enter first number: "))
number2 = float(input("Enter second number: "))
op = input("Enter operator (+, -, *, /): ")

result = calculator(number1, number2, op)
print("Result", result)
    
```

```python
def add_to_list(my_list, value_to_append, value_to_insert, index):
    my_list.append(value_to_append)
    my_list.insert(index, value_to_insert)
    return my_list

my_list = add_to_list(my_list, 6, 5, 0)
print(my_list)
```

```python
def remove_from_list(my_list, value_to_remove):
    if my_list:
        my_list.pop()
    if value_to_remove in my_list:
        my_list.remove(value_to_remove)
    return my_list

my_list = remove_from_list(my_list, 5)
print(my_list)
```

```python
def my_function(*numbers):
    total = 0
    for number in numbers:
        total += number
    return total
print(my_function(1,2,3))
print(my_function(4,5,6,8))
print(my_function(10,20,30,40,50))
```

```python
def my_function(*numbers):
    if len(numbers) == 0:
        return None
    max_num = numbers[0]
    for number in numbers:
        if number > max_num:
            max_num = number
    return max_num
print(my_function(3,5,2,8,1))
```

```python
import random
secrete = random.randint(1, 10)
guess = int(input("Guess a number between 1 and 10: "))
while guess != secrete:
    if guess < secrete:
        print("Too low! Try again.")
    else:
        print("Too high! Try again. ")
    guess = int(input("Guess a number between 1 and 10: "))
print("Congratulations! You've guessed the correct number:", secrete)
```

```python
#create calculator using function
def calculator(number1, number2, operator):
    if operator == "+":
        return number1 + number2
    elif operator == "-":
        return number1 - number2
    elif operator == "*":
        return number1 * number2
    elif operator == "/":
        if number2 == 0:
            return "Error: Cannot divide by Zero"
        return number1 / number2
    else:
        return "Enter a valid operator"
        
number1 = int(input("Input first number: "))
number2 = int(input("Input second number: "))
operator = input("Input operator: ")

result = calculator(number1, number2, operator)
print("Result: ", result)
        
```
