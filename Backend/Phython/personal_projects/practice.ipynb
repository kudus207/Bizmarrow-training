```python
# build_a_caesar_cypher
def caesar(text, shift, encrypt=True):
    # Define the main Caesar cipher function
    # text: the message to encrypt or decrypt
    # shift: how many letters to shift
    # encrypt: True = encrypt, False = decrypt

    if not isinstance(shift, int):
        # Check that shift is an integer
        # If not, stop execution and return an error message
        return 'Shift must be an integer value.'

    if shift < 1 or shift > 25:
        # Ensure shift is within the valid Caesar cipher range
        # (shifting by 0 or 26 would not change anything)
        return 'Shift must be an integer between 1 and 25.'

    alphabet = 'abcdefghijklmnopqrstuvwxyz'
    # Base alphabet used for shifting

    if not encrypt:
        # If we are decrypting (encrypt is False),
        # reverse the direction of the shift
        shift = -shift
    
    shifted_alphabet = alphabet[shift:] + alphabet[:shift]
    # Rotate the alphabet using slicing
    # Example (shift = 3):
    # alphabet[3:]  -> defghijklmnopqrstuvwxyz
    # alphabet[:3]  -> abc
    # result        -> defghijklmnopqrstuvwxyzabc

    translation_table = str.maketrans(
        alphabet + alphabet.upper(),
        shifted_alphabet + shifted_alphabet.upper()
    )
    # Create a translation table:
    # - maps lowercase letters to shifted lowercase letters
    # - maps uppercase letters to shifted uppercase letters

    encrypted_text = text.translate(translation_table)
    # Apply the translation table to the input text

    return encrypted_text
    # Return the encrypted or decrypted text


def encrypt(text, shift):
    # Wrapper function for encryption
    # Calls caesar() with encrypt=True (default)
    return caesar(text, shift)
    

def decrypt(text, shift):
    # Wrapper function for decryption
    # Calls caesar() with encrypt=False
    return caesar(text, shift, encrypt=False)


encrypted_text = 'Pbhentr vf sbhaq va hayvxryl cynprf.'
# This is an encrypted message (ROT13 cipher)

decrypted_text = decrypt(encrypted_text, 13)
# Decrypt the message using a shift of 13

print(decrypted_text)
# Output the decrypted result
```

```python
# Build an RPG Character
def create_character(name, strength, intelligence, charisma):
    # ---------- NAME VALIDATION ----------
    
    # Check if the name is a string
    # isinstance() checks the data type
    if not isinstance(name, str):
        return "The character name should be a string"
    
    # Check if the name is empty
    if name == "":
        return "The character should have a name"
    
    # Check if the name is longer than 10 characters
    if len(name) > 10:
        return "The character name is too long"
    
    # Check if the name contains spaces
    if " " in name:
        return "The character name should not contain spaces"
    
    # ---------- STATS VALIDATION ----------
    
    # Store stats in a list for easy checking
    stats = [strength, intelligence, charisma]
    
    # Check if all stats are integers
    # all() ensures every stat passes the check
    if not all(isinstance(stat, int) for stat in stats):
        return "All stats should be integers"
    
    # Check if any stat is less than 1
    if not all(stat >= 1 for stat in stats):
        return "All stats should be no less than 1"
    
    # Check if any stat is greater than 4
    if not all(stat <= 4 for stat in stats):
        return "All stats should be no more than 4"
    
    # Check if total stat points equal 7
    if sum(stats) != 7:
        return "The character should start with 7 points"
    
    # ---------- OUTPUT FORMATTING ----------
    
    # Helper function to format each stat line
    def format_stat(label, value):
        # Full dots represent the stat value
        # Empty dots fill the remaining space to make 10
        return label + " " + "●" * value + "○" * (10 - value)
    
    # Return the final character description
    return (
        name + "\n" +
        format_stat("STR", strength) + "\n" +
        format_stat("INT", intelligence) + "\n" +
        format_stat("CHA", charisma)
    )

print(create_character("", 4, 2, 1))
print(create_character("verylongcharactername", 2, 2, 3))
print(create_character("hero", 3, 3, 1))
```

```python
# Build a pin extractor
def pin_extractor(poems):
    # Step 1: Create an empty list to store all extracted PINs
    secret_codes = []

    # Step 2: Loop through each poem in the list of poems
    for poem in poems:
        # Step 3: Create an empty string to build the PIN for this poem
        secret_code = ''

        # Step 4: Split the poem into individual lines using newline characters
        lines = poem.split('\n')

        # Step 5: Loop through each line with its index
        for line_index, line in enumerate(lines):
            # Step 6: Split the current line into words
            words = line.split()

            # Step 7: Check if there is a word at the current index
            if len(words) > line_index:
                # Step 8: Add the length of that word to the PIN
                secret_code += str(len(words[line_index]))
            else:
                # Step 9: If the word does not exist, add '0' to the PIN
                secret_code += '0'

        # Step 10: After processing all lines, save the PIN for this poem
        secret_codes.append(secret_code)

    # Step 11: Return the list of all extracted PINs
    return secret_codes


# Poems to extract PINs from
poem = """Stars and the moon
shine in the sky
white and
until the end of the night"""

poem2 = 'The grass is green\nhere and there\nhoping for rain\nbefore it turns yellow'
poem3 = 'There\nonce\nwas\na\ndragon'

# Step 12: Call the function with multiple poems and print the result
print(pin_extractor([poem, poem2, poem3]))
```

```python
# Build a number pattern generator
def number_pattern(n):
    # Check if the argument is an integer
    if not isinstance(n, int):
        return "Argument must be an integer value."
    
    # Check if the integer is greater than 0
    if n < 1:
        return "Argument must be an integer greater than 0."
    
    # Initialize an empty string to store the result
    result = ""
    
    # Use a for loop to build the number pattern
    for i in range(1, n + 1):
        result += str(i)
        
        # Add a space if it's not the last number
        if i != n:
            result += " "
    
    return result
print(number_pattern(1))      
print(number_pattern(4))      
print(number_pattern(0))     
print(number_pattern("4"))    
```
